This is a Divvy Bike (bike share) data set from Chicago. It is downloaded from [Chicago Data Portal](https://data.cityofchicago.org/Transportation/Divvy-Trips/fg6s-gzvg/data). 

The csv in the data folder has been filtered to look at only 1 pickup station: Clark and Lake. That is a busy intersection in the Loop.  

```python
# Import the dataset

df = None
```

```python
#__SOLUTION__
# Raw Dataset
import pandas as pd
df = pd.read_csv('data/Divvy_Trips.csv')
df.head()
```

We will be interested in the number of Divvy rentals over a given time period.

To begin, we will look at it by the **hour**.  

In order to begin inspecting the data across different time intervals, we need to give our dataset a datetime index.

```python
# Create datetime index based on the start time
```

```python
#__SOLUTION__
# Create datetime index based on the start time

df['START TIME'] = pd.to_datetime(df['START TIME'])
df.set_index(df['START TIME'], inplace=True)
```

Now that we have a datatime index, we can resample to a time frame of our choosing.

We are going from a smaller unit of time to a larger unit, so we will downsample.  With any resampling technique, we need to specify an aggregate function. In this case, we want to count the number of rides per hour.

```python
# Downsample to the hour and count the number of rides
```

```python
#__SOLUTION__
# Downsample to the hour and count the number of rides
df_hour = df.resample('H').count()
df_hour
```

```python
df_hour['TRIP ID'].plot()
```

That is a dense plot. Let's look at what's happening over the course of a week.

```python
df_hour['TRIP ID'][:24*7].plot()
```

In order to see any trends, let's plot some rolling statistics. 

```python
df_hour['TRIP ID'].rolling(24*7).mean().plot()
```

```python
df_hour['TRIP ID'].rolling(24*7*52).mean().plot()
```

```python
# Downsample to Daily
df_daily = df.resample('D').count()['TRIP ID']
df_daily.plot()
```

```python
# Train test split
train_daily = df_daily[:round(len(df_daily)*.8)]
test_daily = df_daily[round(len(df_daily)*.8):]
```

```python
train_daily.plot()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(train_daily)
ax.plot(train_daily.rolling(365).mean())
ax.plot(train_daily.rolling(365).std())
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(df_daily.diff().dropna());
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

sm = SARIMAX(train_daily, order = [0,1,0]).fit()
```

```python
sm.predict(typ='levels')
```

```python
from sklearn.metrics import mean_squared_error
rmse = mean_squared_error(train_daily, sm.predict(typ='levels'), squared=False)
rmse
```

```python
def print_ts_metrics( order, endog=train_daily, seasonal_order=[0,0,0,0]):
    sm = SARIMAX(endog, order = order, seasonal_order=seasonal_order ).fit()

    print('AIC: ', sm.aic)
    rmse = mean_squared_error(endog, sm.predict(typ='levels'), squared=False)
    print('rmse: ',  rmse)
    
    return sm

print_ts_metrics([1,1,1])
```

```python
print_ts_metrics([1,2,1],train_daily )
```

```python
print_ts_metrics([2,2,1], train_daily)
```

```python
print_ts_metrics([1,1,1], train_daily, [1,1,1,7], )
```

```python
def plot_predictions(test=test_daily, sm=sm):
    sm.predict(test.index[0], test.index[-1], typ='levels').plot()
    
```

```python
model = print_ts_metrics([1,1,1], train_daily, [1,1,1,7])
plot_predictions(test_daily, model)
```

```python
model = print_ts_metrics([1,1,1],train_daily,  [1,1,1,7])
plot_predictions(test_daily, model)
test_daily.plot(color='r')
```

```python
model = print_ts_metrics([1,1,1],train_daily.diff(365).dropna(),  [1,1,1,7])
plot_predictions(test_daily.diff(365), model)
test_daily.plot(color='r')
```

```python
model = print_ts_metrics([1,1,1],train_daily,  [1,1,1,365])
plot_predictions(test_daily, model)
test_daily.plot(color='r')
```

```python
model = print_ts_metrics([1,1,1],train_daily,  [1,1,1,7])
plot_predictions(test_daily, model)
```

```python
model = print_ts_metrics([1,1,1],train_daily,  [1,1,1,52])
plot_predictions(test_daily, model)
```
